```python
# GitHub Profile QR Code Generator
# This project creates a QR code that links directly to a GitHub profile
# Purpose: Generate a scannable QR code image for easy profile sharing

# Step 1: Install and import required libraries
!pip install qrcode
import qrcode
from PIL import Image

# Step 2: Configure QR code settings
  # Version 1: Smallest QR code size
qr = qrcode.QRCode(  # High error correction: QR code remains readable even if partially damaged
    version=1,  # Box size: Size of each module/square in pixels
    error_correction=qrcode.constants.ERROR_CORRECT_H,  # Border: Width of the border (in boxes)
    box_size=10,
    border=4,
)
```

```python
qr.add_data("https://github.com/Mariam-iftikhar")
qr.make(fit=True)
img = qr.make_image(fill_color="white", back_color="blue")

# Step 3: Generate the QR code with GitHub profile URL
img.save("github_link.png")  # Add the GitHub profile URL to the QR code

 # Finalize the QR code with optimal fit
 # Create QR code image with white modules on blue background

 # Save the QR code as 'github_link.png' in the current directory
```
