```python
# PDF Splitter Tool
# This project splits a multi-page PDF into individual single-page PDF files
# Purpose: Extract each page from a PDF and save it as a separate PDF file

# Step 1: Install and import the required library

!pip install pikepdf  # Install pikepdf - a Python library for PDF manipulation

import pikepdf  # Import the pikepdf library

old_pdf = pikepdf.Pdf.open("/content/sample_data/testfile.pdf")  # Open the PDF file from the specified path
```

```python
# Step 2: Split the PDF into individual pages

for n, page_can in enumerate(old_pdf.pages):  # Loop through each page in the PDF
    new_pdf = pikepdf.Pdf.new()  # Create a new empty PDF document
    new_pdf.pages.append(page_can)  # Add the current page to the new PDF
    name = "test" + str(n) + ".pdf"  # Generate filename: 'test1.pdf', 'test2.pdf', etc.
    new_pdf.save(name)  # Save the single-page PDF with the generated filename
```
